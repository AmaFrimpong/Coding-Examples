# Creating age bins and plotting the survival rate for each age group

```python
age_bins = [0, 3, 10, 18, 30, 60, 90]
```

```python
bin_labels = ['baby', 'toddler', 'teen', 'adult' 'middleleaged', 'older']
```

#### Need fewer bins than bin count

```python
age_bin_series = pd.cut(df.Age, age_bins, bin_labels,)
```

```python
df['Age Bin'] = age_bin_series
```

```python
df.groupby(['Age Bin', 'Survived']).count()
```

```python
res = df.groupby(['Age Bin', 'Survived']).size().unstack()
res.columns = [['Not Survived', 'Survived']]
res.plot(kind='bar')
```

```python
sdict = {0: 'Not Survived', 1: 'Survived'}
```

```python
sdict
```

```python
df['Survived Desc'] = df['Survived'].map(sdict) #Takes the values in, maps them to dictionary and returns them
```

```python
df
```
