```python
import urllib.request
import json
from pprint import pprint
import pandas as pd
from pandas.io.json import json_normalize
```

```python
url = r'http://datapoint.metoffice.gov.uk/public/data/val/wxobs/all/json/3772?res=hourly&key=acea5d3c-ec7d-49ad-9828-29f485a93181'
```

```python
with urllib.request.urlopen(url) as response:
    html = response.read()
    htmld = json.loads(html)
    pprint(htmld)
```

```python
#The first parts that are bracketed (before the ,) is
# the sliced parts then just before the data you want, add a comma
# in order to display the data within 'Rep'
weather = json_normalize(htmld['SiteRep']['DV']['Location']['Period'], ['Rep'], ['value', 'type'])
```

```python
weather
```

# How to create a column which shows if the time is daylight or not.

```python
# The hour column labelled $, so we want to extract it
```

```python
t = weather[['$']]
#The hour column is encased in 2 square brackets
# to show it is a DataFrame
```

```python
# Imagine that this t is just the $, it was re-run after the daylight
#if statement
t
```

```python
# t in daylight is what we want, and we are mapping every item in $ to
# a new function i.e. daylight
t['daylight'] = t['$'].map(daylight)
```

```python
t['workhours'] = t['$'].map(lambda x: 1 if int(x) >= 540 and int(x) <= 960 else 0)
```

```python
def daylight(m):
    if int(m) >= 420 and int(m) <= 960:
        r = 1
    else:
        r = 0
    return r
```

```python
t.drop('hours', 1)
```

```python
t['hours'] = t['$'].map(lambda x: int(x)/60)
```

### Joining tables
